Short theory: when to use mean, median, mode?

* **Numeric, roughly symmetric, few outliers  *mean***

  * Pros: uses all values; Cons: pulled by outliers.
* **Numeric, skewed or with outliers  *median***

  * Robust to outliers; preserves central tendency for skewed data.
* **Ordinal (ordered categories like low/med/high)  *median*** (after mapping to ordered integers).
* **Nominal categorical (e.g., Embarked)  *mode*** (most frequent).
* **Time series gaps  *forward/backward fill*** (`ffill`/`bfill`) or interpolation, but beware of leaking future info.
* **Structural zeros or meaningful missing  constant + missing flag**


**Rules of thumb**

* Inspect distribution first (hist/boxplot). If skewed/outliers  prefer **median**.
* For categoricals  **mode** (or create Unknown if missingness carries signal).
* Consider **group-wise** imputation (by `Sex`, `Pclass`, etc.) to respect heterogeneity.
* In ML, do imputation **inside the training pipeline** to avoid leakage and keep transforms identical at inference.

#### Modeling: Decision Tree

For this classification problem, we will use a **Decision Tree**, a simple yet powerful and highly interpretable model.

A Decision Tree works by **sequentially splitting the dataset into smaller and purer groups** based on the values of different features. At each step, the algorithm automatically selects the **feature and split point** that best separates the classes  for example, splitting passengers by Sex first, then Pclass, and so on.

You can think of a Decision Tree as a **series of yes/no questions** that progressively narrow down the possibilities until a final decision is reached. For example:

> Is the passenger female?  Yes  Is the fare > 30?  Yes  likely survived.

This structure makes the model:

* **Intuitive and easy to visualize**  its logic is similar to human decision-making processes.
* **Capable of handling both numerical and categorical variables** with minimal preprocessing.
* **Flexible**, but potentially prone to overfitting if grown too deep, which is why we control complexity using parameters like `max_depth` and `min_samples_leaf`.

In the context of the Titanic dataset, we **expect certain features to strongly influence the trees splitting decisions** (see the previous plots).

* **Sex** is a key factor because survival rates were significantly higher for female passengers due to evacuation priorities (women and children first).
* **Pclass** (passenger class) encodes socio-economic status, and in this dataset, higher classes had better access to lifeboats and therefore higher survival probabilities.

Because Decision Trees naturally prioritize **features that best reduce impurity**, it is likely that **Sex** and **Pclass** will appear near the **top of the tree**, driving the most influential early splits. This also illustrates how trees can **surface strong biases present in the data**  in this case, reflecting real historical inequalities during the Titanic disaster.

In this lab, we will train a basic Decision Tree classifier as our first model. In future sessions, well explore in more detail how trees choose splits, how to tune their complexity, and how they serve as the foundation for advanced ensemble methods like **Random Forests** and **Gradient Boosted Trees**.

To evaluate how well a model performs on **unseen data**, the dataset is typically divided into separate subsets for training, validation, and testing.

* The **training set** is used to fit the models parameters.
* The **validation set** supports model selection and hyperparameter tuning (more about this in the next laboratories).
* The **test set** provides an unbiased estimate of the models ability to **generalize**, i.e., to make accurate predictions on data it has never encountered before.

For this evaluation to be meaningful, the subsamples must follow the **same underlying distribution** as the full dataset. In classification problems, this often means preserving class proportions (stratification). In other contexts, such as time series or grouped data, special splitting strategies are needed to avoid information leakage between training and testing subsets.

Data leakage must be avoided at all stages: any transformation or preprocessing should be learned exclusively from the training set and then applied to other subsets. Otherwise, the model may indirectly see information from the test data, leading to overly optimistic results.

Finally, model performance should not be judged on a single split alone. In practice, **cross-validation** or multiple validation rounds are often used to obtain more stable performance estimates and better model selection. The test set is reserved for the final performance check after all decisions have been made.
